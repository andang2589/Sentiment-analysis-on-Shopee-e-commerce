# Crawl

```python
comment=[]

for i in range(0,1000):
  url= 'https://shopee.vn/api/v2/item/get_ratings?filter=0&flag=1&itemid=4020345082&limit=6&offset='+ str(i*6)+'&shopid=66986483&type=0' 
  headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
  r=requests.get(url,headers=headers)
  soup=r.json()
  data=soup['data']['ratings']
  for review in data:
    if review['rating_star']==1:
      comment.append(review['comment'])
```

```python
rating={'Comment':comment}
```

```python
df_shoppe=pd.DataFrame(rating)
df_shoppe
```
