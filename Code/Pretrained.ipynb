```python
def convert_unicode(cmt):
    cmt = unicodedata.normalize('NFC', cmt)
    return cmt
```

```python
def comment_preprocess(cmt):
    # chu·∫©n h√≥a unicode
    cmt = convert_unicode(cmt)
    # chu·∫©n h√≥a c√°ch g√µ d·∫•u ti·∫øng Vi·ªát (√≤a -> o√†, √∫y -> u√Ω,...) v√† t√°ch t·ª´
    cmt = rdrsegmenter.tokenize(cmt)
    #cmt = ' '.join(' '.join(word for word in sentence) for sentence in cmt)
    # ƒë∆∞a v·ªÅ lower
    cmt = cmt.lower()
    # x√≥a c√°c k√Ω t·ª± kh√¥ng c·∫ßn thi·∫øt
    cmt = re.sub(r'[^\s\w√°√†·∫£√£·∫°ƒÉ·∫Ø·∫±·∫≥·∫µ·∫∑√¢·∫•·∫ß·∫©·∫´·∫≠√©√®·∫ª·∫Ω·∫π√™·∫ø·ªÅ·ªÉ·ªÖ·ªá√≥√≤·ªè√µ·ªç√¥·ªë·ªì·ªï·ªó·ªô∆°·ªõ·ªù·ªü·ª°·ª£√≠√¨·ªâƒ©·ªã√∫√π·ªß≈©·ª•∆∞·ª©·ª´·ª≠·ªØ·ª±√Ω·ª≥·ª∑·ªπ·ªµƒë_]',' ',cmt)
    # x√≥a kho·∫£ng tr·∫Øng th·ª´a
    cmt = re.sub(r'\s+', ' ', cmt).strip()
    return cmt
```

```python
def remove_stopwords(line):
    global stopword_count
    words = []
    for word in line.strip().split():
        if word not in stopword:
            words.append(word)
        if word in stopword:
            stopword_count += 1
    return ' '.join(words)
```

```python
# Ti·ªÅn x·ª≠ l√Ω
from tensorflow.keras.preprocessing.sequence import pad_sequences
from pyvi import ViTokenizer
from keras.preprocessing.text import Tokenizer

max_len = 300

word_tokenizer = Tokenizer(oov_token=-1)
word_tokenizer.fit_on_texts(X_train)
word_to_index = word_tokenizer.word_index
word_to_index['pad'] = 0
word_to_index['unk'] = -1

index_to_word = {i: w for w, i in word_to_index.items()}

def encoding(X):
    sentences = []
    
    for t in X:
        tokenized_sentence = ViTokenizer.tokenize(t)
        sentences.append(tokenized_sentence)
    
    X = word_tokenizer.texts_to_sequences(sentences)
    X = pad_sequences(maxlen = max_len, sequences = X, padding = "post", value = word_to_index['pad'])

    return X
```

```python
word_to_index
```

```python
import numpy as np 
# first create a matrix of zeros, this is our embedding matrix
num_words = len(word_to_index)

embedding_matrix = np.zeros((num_words, embedding_dim))

# for each word in out tokenizer lets try to find that work in our w2v model
for word, i in word_to_index.items():
    if i > max_feature:
        continue
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
    else:
        # doesn't exist, assign a random vector
        embedding_matrix[i] = np.random.randn(embedding_dim)
```

```python
# Encoding 
X_train_new=encoding(X_train)
X_test_new=encoding(X_test)
```

```python
# X√¢y d·ª±ng m·∫°ng neural
from keras.layers import Dense, Embedding, Flatten
from keras.models import Model, Input
from keras.initializers import Constant

num_words = len(word_to_index)

input = Input(shape = (max_len, ))
emb = Embedding(input_dim=num_words,
                    output_dim=embedding_dim,
                    embeddings_initializer=Constant(embedding_matrix),
                    input_length=max_len,
                    trainable=True)(input)
flat = Flatten()(emb)
output = Dense(3, activation="sigmoid")(flat)

model = Model(input, output)
model.compile(optimizer="adam", loss='binary_crossentropy', metrics=['binary_accuracy'])

model.summary()
```

```python
from tensorflow.keras.utils import to_categorical

history= model.fit(X_train_new, to_categorical(y_train, num_classes=3), batch_size=128, epochs=10)
```

```python
import numpy as np
y_pred = model.predict(X_test_new)
y_pred = np.argmax(y_pred, axis=-1)
```
